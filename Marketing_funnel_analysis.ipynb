```python
import pandas as pd
import matplotlib.pyplot as plt

# Load users data
df = pd.read_csv("../Data/user_table.csv")

df.head()
```

```python
df.info()
```

```python
# Load funnel stage tables

home = pd.read_csv("../Data/home_page_table.csv")
search = pd.read_csv("../Data/search_page_table.csv")
payment = pd.read_csv("../Data/payment_page_table.csv")
confirmation = pd.read_csv("../Data/payment_confirmation_table.csv")

# Funnel counts

funnel_counts = {
    "Home Page": len(home),
    "Search Page": len(search),
    "Payment Page": len(payment),
    "Payment Confirmation": len(confirmation)
}

print(funnel_counts)
```

```python
# Funnel visualization

stages = list(funnel_counts.keys())
counts = list(funnel_counts.values())

plt.figure(figsize=(8,5))

plt.bar(stages, counts)

plt.title("Marketing Funnel Analysis")
plt.xlabel("Funnel Stage")
plt.ylabel("Users")

plt.xticks(rotation=10)

plt.show()
```

```python
# Device distribution

device_counts = df['device'].value_counts()

print(device_counts)
```

```python
device_counts.plot(kind='bar')

plt.title("User Device Distribution")
plt.xlabel("Device")
plt.ylabel("Users")

plt.xticks(rotation=0)

plt.show()
```

```python
Key Insights:

1. Significant user drop-off occurred between funnel stages.
2. Only a smaller percentage of users reached payment confirmation.
3. Desktop users contributed major platform traffic.
4. Funnel optimization can improve customer conversion rates.
5. Businesses should focus on reducing drop-offs during payment stages.
```

```python
# Conversion Rate Calculation

total_users = len(home)
final_conversions = len(confirmation)

conversion_rate = (final_conversions / total_users) * 100

print("Overall Conversion Rate:", round(conversion_rate, 2), "%")
```

```python
# Gender distribution

gender_counts = df['sex'].value_counts()

print(gender_counts)
```

```python
gender_counts.plot(kind='bar')

plt.title("User Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Users")

plt.xticks(rotation=0)

plt.show()
```

```python
# Convert date column

df['date'] = pd.to_datetime(df['date'])

daily_users = df.groupby('date')['user_id'].count()

print(daily_users.head())
```

```python
daily_users.plot(figsize=(12,5))

plt.title("Daily Website Traffic")
plt.xlabel("Date")
plt.ylabel("Users")

plt.show()
```

```python
Final Business Insights:

1. Large user drop-offs occurred before payment confirmation.
2. Funnel optimization can improve conversion rates significantly.
3. Desktop users generated higher platform traffic.
4. Monitoring daily traffic trends helps businesses track customer engagement.
5. Businesses should improve checkout experience to reduce customer abandonment.
```

```python
# Funnel stage counts

home_users = len(home)
search_users = len(search)
payment_users = len(payment)
confirmed_users = len(confirmation)

print("Home Page Users:", home_users)
print("Search Page Users:", search_users)
print("Payment Page Users:", payment_users)
print("Confirmed Users:", confirmed_users)
```

```python
# Drop-off calculations

search_drop = ((home_users - search_users) / home_users) * 100

payment_drop = ((search_users - payment_users) / search_users) * 100

confirmation_drop = ((payment_users - confirmed_users) / payment_users) * 100

print("Drop-off Home → Search:", round(search_drop,2), "%")
print("Drop-off Search → Payment:", round(payment_drop,2), "%")
print("Drop-off Payment → Confirmation:", round(confirmation_drop,2), "%")
```

```python
# Funnel chart

stages = ['Home', 'Search', 'Payment', 'Confirmed']
values = [home_users, search_users, payment_users, confirmed_users]

plt.figure(figsize=(8,5))

plt.plot(stages, values, marker='o')

plt.title("Customer Funnel Drop-Off")
plt.xlabel("Stages")
plt.ylabel("Users")

plt.show()
```

```python
Advanced Funnel Insights:

1. Major customer drop-offs occurred between search and payment stages.
2. A smaller percentage of users completed final payment confirmation.
3. Funnel optimization opportunities exist in checkout flow.
4. Businesses can improve conversions by simplifying payment experience.
5. Monitoring customer journey stages helps improve revenue generation.
```
